Matplotlib:
  useful for making plots

```python
#importing matplot library
import matplotlib.pyplot as plt
```

```python
#import numpy
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x,y,z)
```

```python
print(y)
```

```python
print(z)
```

plotting the data

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel("angel")
plt.ylabel("sine value")
plt.title("sinwave")
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show
```

```python
x = np.linspace(-5,5,70)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r+')
plt.show()
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = [ 'english','tamil','spanish','french','hindi']
people = [100,50,30,20,10]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = [ 'english','tamil','spanish','french','hindi']
people = [100,50,30,20,10]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot


useful in clustering applications

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="b")
plt.show()
```

3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
